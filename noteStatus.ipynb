## Note Status History


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
stat_data = pd.read_csv('data/noteStatusHistory.csv')
```

### Pie Chart for Note Status
This pie chart compares the amount of notes who have received a status of either helpful or not helpful. As of right now, more notes are currently rated as helpful as compared to the amount of notes that are rated as not helpful. However, it can be seen in the second pie chart that the vast majority of tweets still need more ratings to be deemed as either helpful or not helpful.

```python
# Count the occurrences of 'believable' values
stat_count = stat_data[stat_data['currentStatus'] != 'NEEDS_MORE_RATINGS']['currentStatus'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
ax = stat_count.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart for Note Status')
plt.show()
```

```python
# Count the occurrences of 'believable' values
stat_count = stat_data['currentStatus'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
ax = stat_count.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart for Note Status')
plt.show()
```

### Models used to decide Ratings
This bar graph shows the amount of tweets that use a certain algorithm to decide on the rating that they receive. It can be seen that most tweets use CoreModel (v1.1).

```python
counts = stat_data['currentDecidedBy'].value_counts()

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
ax = counts.plot(kind='bar', x='Combination', y='Count', stacked=True, color=['lightblue'])
plt.xlabel('Model')
plt.ylabel('Count')
plt.title('Models used to decide Ratings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
